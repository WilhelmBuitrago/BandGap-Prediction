```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import ast
import time
from IPython.display import clear_output
```

```python
df_fluorite = pd.read_csv('Dataset\Bandgap\dataset_Paper1_Fluorite.csv', index_col="ID")
df_fluorite['x1'] = df_fluorite['x1'].apply(ast.literal_eval)
df_fluorite['x2'] = df_fluorite['x2'].apply(ast.literal_eval)

df_Mukesh = pd.read_csv('Dataset\Bandgap\dataset_Paper2_Mukesh.csv', index_col="ID")
df_Mukesh['x1'] = df_Mukesh['x1'].apply(ast.literal_eval)
df_Mukesh['x2'] = df_Mukesh['x2'].apply(ast.literal_eval)
```

```python
id = 3
print(len(df_Mukesh.iloc[id][0]))
print(len(df_fluorite.iloc[id][0]))
print(df_Mukesh.iloc[id].iloc[0][-1])
print(df_fluorite.iloc[id].iloc[0][0]-df_fluorite.iloc[id].iloc[0][-1])
```

```python
def graficar_5(df, id):
    fig, axs = plt.subplots(2,3, figsize=(9, 6)) 

    axs[0,0].plot(df.iloc[id].iloc[0], df.iloc[id].iloc[1])
    axs[0,0].scatter(df.iloc[id].iloc[2], 0, color='red', label=f'Bandgap: {df.iloc[id].iloc[2]}')
    axs[0,0].set_title(f'Gráfica ID: {id} Bandgap: {df.iloc[id].iloc[2]}')

    axs[0,1].plot(df.iloc[id+1].iloc[0], df.iloc[id+1].iloc[1])
    axs[0,1].scatter(df.iloc[id+1].iloc[2], 0, color='red')
    axs[0,1].set_title(f'Gráfica ID: {id+1} Bandgap: {df.iloc[id+1].iloc[2]}')

    axs[0,2].plot(df.iloc[id+2].iloc[0], df.iloc[id+2].iloc[1])
    axs[0,2].scatter(df.iloc[id+2].iloc[2], 0, color='red')
    axs[0,2].set_title(f'Gráfica ID: {id+2} Bandgap: {df.iloc[id+2].iloc[2]}')

    axs[1,0].plot(df.iloc[id+3].iloc[0], df.iloc[id+3].iloc[1])
    axs[1,0].scatter(df.iloc[id+3].iloc[2], 0, color='red')
    axs[1,0].set_title(f'Gráfica ID: {id+3} Bandgap: {df.iloc[id+3].iloc[2]}')

    axs[1,1].plot(df.iloc[id+4].iloc[0], df.iloc[id+4].iloc[1])
    axs[1,1].scatter(df.iloc[id+4].iloc[2], 0, color='red')
    axs[1,1].set_title(f'Gráfica ID: {id+4} Bandgap: {df.iloc[id+4].iloc[2]}')

    plt.tight_layout()
    plt.show()

def graficar(df, id):
    x = np.array(df_fluorite.iloc[id].iloc[0])
    y = np.array(df_fluorite.iloc[id].iloc[1])
    # Graficar los datos para el ID actual
    plt.plot(x, y, label='UV-Vis')
    plt.scatter(df_fluorite.iloc[id].iloc[2], 0, color='red', label=f'Bandgap: {df_fluorite.iloc[id].iloc[2]}')
    plt.xlabel('Absorbancia')
    plt.ylabel('Frecuencia')
    plt.title(f'Gráfica para ID {id}')
    plt.legend()
    plt.show()
```

```python
graficar(df_fluorite,529)
```

```python
graficar_5(df_Mukesh,0)
```

```python
for id in range(530):
    graficar(df_fluorite,id)
    time.sleep(2)
    plt.close()
    clear_output(wait=True)
```

```python
for id in range(0,530,5):
    # Filtrar el DataFrame para obtener solo las filas correspondientes al ID actual
    graficar_5(df_Mukesh,id)
    time.sleep(5)
    plt.close()
    clear_output(wait=True)
```

```python
import keyboard
ID = -5
on = True
def accion_tecla(event):
    global ID
    global on
    if event.name == 'right':
        ID += 5
        plt.close()
        clear_output(wait=True)
        graficar(ID)
    elif event.name == 'left':
        ID -= 5
        plt.close()
        clear_output(wait=True)
        graficar(ID)
    elif event.name == 'esc':
        on = False

keyboard.on_press(accion_tecla)

while on: 
    pass
```
